```python
from datetime import date, datetime
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt

start_date = '2011-11-30'
end_date = '2021-11-30'
```

```python
series = [NVDA,VERIZON,ATT,AMC,WMT,COST]
#stampa punto 2.d

    
def diagnostics(serie):
    rtn = serie['Adj Close']/serie['Adj Close'].shift(1)
    plt.figure()
    plt.hist(rtn, density=True)
    rtn.plot.density()
    
    plt.figure()
    plt.boxplot(rtn, sym = 'x')
    plt.title("Boxplot")
    rtn.to_frame().boxplot()
    
    import statsmodels.api as sm
    sm.qqplot(rtn, line = 's')
    
    import seaborn as sns
    sns.set_style('whitegrid')
    sns.kdeplot(np.array(rtn))

for x in series:
    diagnostics(x)
```

```python
series = [NVDA,VERIZON,ATT,AMC,WMT,COST]
#stampo punto 2.e
def descUnivariate(serie):
    rtn = serie['Adj Close']/serie['Adj Close'].shift(1)
    
    print(round(rtn.mean(), 4))
    print(round(rtn.var(), 4))
    print(round(rtn.std(), 4))
    print(round(rtn.skew(), 4))
    print(round(rtn.kurtosis(), 4))
    
    print("--------------")
    
for x in series:
    descUnivariate(x)
```
